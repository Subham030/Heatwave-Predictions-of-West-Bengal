```python
import os
import pandas as pd
import numpy as np

def process_csv(file_path, output_dir, n, t):
  
    data = pd.read_csv(file_path)

    
    data['time'] = data['hour']

    values_utci = data['UTCI_Celsius_mean'].values
    X, y = [], []

    for i in range(len(values_utci) - n - t + 1):
        X.append(values_utci[i:i + n])
        y.append(values_utci[i + n:i + n + t])

    
    os.makedirs(output_dir, exist_ok=True)

    
    base_name = os.path.basename(file_path).replace('.csv', '')
    np.save(os.path.join(output_dir, f"{base_name}_features.npy"), np.array(X))
    np.save(os.path.join(output_dir, f"{base_name}_targets.npy"), np.array(y))

def process_directory(input_dir, output_dir, n, t):
    
    os.makedirs(output_dir, exist_ok=True)

    for file_name in os.listdir(input_dir):
        if file_name.endswith('.csv'):
            file_path = os.path.join(input_dir, file_name)
            process_csv(file_path, output_dir, n, t)

print("done")


input_directory = 'data_more_1990-2002'  
output_directory = 'data_more_5-1_array_1990-2002'  
n_past = 5
t_future = 1  

process_directory(input_directory, output_directory, n_past, t_future)
```
